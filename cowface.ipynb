```python
import torch
from os import makedirs, listdir
from os.path import isdir
import subprocess
from cv2 import imread, imwrite
```

```python
model = torch.hub.load(repo_or_dir='ultralytics/yolov5', model='custom', path='best.pt')
```

```python
def run_inference(img, dest):
    results = model(img)
    results.save(save_dir=dest)
```

```python
frames = sorted((f for f in listdir('/content/demo/frames/norm') if not f.startswith(".") and not isdir(f'/content/demo/frames/norm/{f}')),
                key=str.lower)
for f in frames:
    run_inference(f'/content/demo/frames/norm/{f}', f'/content/demo/frames/norm/boxes')
```
